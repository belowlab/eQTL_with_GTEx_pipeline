Run eQTL mapping using batch 1-4 RNAseq data, following steps in [GTEx QTL pipeline](https://github.com/broadinstitute/gtex-pipeline/tree/master/qtl)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import os

print('Last run:', datetime.datetime.now().strftime('%y-%m-%d'))
```

# 1. Clean TMM and get phenotype covariates

## 1.1 Map RNAseq LABIDs to RRIDs and remove duplicates
1.  Convert LABIDs to genotype IDs
2. Keep the most rescent visit only

```python
fn_tmm = '/vgipiper04/CCHC/RNAseq/batch1_2_3_4/GTEx/expression/CCHC_batch1_2_3_4_TMM.LABID.low_quality_dropped.expression.bed.gz'
df_tmm = pd.read_csv(fn_tmm, sep='\t', compression='gzip')
display(df_tmm.head(2))
print(df_tmm.shape)

fn_pheno = '/data100t1/share/CCHC/phenotypes/1223/cchc_phenotypes_1223.txt'
cols_to_load = ['RRID', 'LABID', 'VISIT', 'AGE_AT_VISIT', 'GENDER']
df_pheno = pd.read_csv(fn_pheno, sep='|', dtype=str)[cols_to_load]
# display(df_pheno.head(2))

# Filter out samples with RNAseq TMMs
print('# Filter samples with RNAseq data on phenotypes')
df_pheno_mapped = df_pheno[df_pheno['LABID'].isin(df_tmm.columns[4:])].copy()
for col in df_pheno_mapped.columns[2:]:
    df_pheno_mapped[col] = pd.to_numeric(df_pheno_mapped[col], errors='coerce')
print(df_pheno_mapped.shape)
display(df_pheno_mapped.head(2))
```

```python
print('# Drop duplicates (keep the last visit)')
n_samples = len(df_pheno_mapped)
df_pheno_mapped.sort_values(by=['RRID', 'VISIT'], inplace=True)
df_pheno_mapped.drop_duplicates(subset=['RRID'], keep='last', inplace=True)
print('# - N samples dropped:', n_samples - len(df_pheno_mapped))
```

```python
# Drop samples without genotype PC
fn_pc = '/data100t1/home/wanying/CCHC/doc/genotype_and_pc/202210_CCHC_PCs_id_fixed_1KG_controls_removed.txt'
df_pc = pd.read_csv(fn_pc, sep='\t')
display(df_pc.head(2))

n_samples = len(df_pheno_mapped)
df_pheno_mapped = df_pheno_mapped.merge(df_pc[['RRID']], on='RRID')
print('# Drop samples without genotype PC')
print('# - N samples dropped:', n_samples - len(df_pheno_mapped))
print(len(df_pheno_mapped))
```

## 1.2 Remove outliers
1. Age between 18 to 80, inclusive
1. Remove samples with missing age and sex: (No missing values in sex and age, 0 samples dropped)

```python
print('# Remove outliers')
mask_age = (df_pheno_mapped['AGE_AT_VISIT']>=18) & (df_pheno_mapped['AGE_AT_VISIT']<=80)
n_samples = len(df_pheno_mapped)
df_pheno_mapped_cleaned = df_pheno_mapped[mask_age].copy()
print('# - N samples dropped by age<18 or >80:', n_samples - len(df_pheno_mapped_cleaned))
print(len(df_pheno_mapped_cleaned))
```

## 1.3 Map filtered RNAseq LABIDs to genotype IDs
Save filtered TMM to a bgzipped file and tabix



```python
fn_genotype_id = '/data100t1/home/wanying/CCHC/doc/samples_IDs/20240325_IDs_genotyped_all.txt'
df_genotype_id = pd.read_csv(fn_genotype_id, sep='\t')
display(df_genotype_id.head(2))

print('# Map filtered RNAseq LABIDs to genotype IDs')
n_samples = len(df_pheno_mapped_cleaned)
df_pheno_mapped_cleaned_id_mapped = df_pheno_mapped_cleaned.merge(df_genotype_id, on='RRID').copy()
print('# - N samples dropped:', n_samples-len(df_pheno_mapped_cleaned_id_mapped))
print('# - Number of sampels left:', len(df_pheno_mapped_cleaned_id_mapped))

print('\n# Convert RNAseq LABID to RRIDs')
cols_filtered = ['#chr', 'start', 'end', 'gene_id'] + list(df_pheno_mapped_cleaned_id_mapped['LABID'])
df_tmm_cleaned = df_tmm[cols_filtered].copy()

# Rename columns from LABID to genotype IDs
# Conver LABIDs to RRIDs
dict_labid2rrid = dict(zip(df_pheno_mapped_cleaned_id_mapped['LABID'], df_pheno_mapped_cleaned_id_mapped['RRID']))
df_tmm_cleaned.rename(columns=dict_labid2rrid, inplace=True)
# Conver RRIDs to genotype IDs
dict_rrid2gpid = dict(zip(df_pheno_mapped_cleaned_id_mapped['RRID'], df_pheno_mapped_cleaned_id_mapped['genotype_ID']))
df_tmm_cleaned.rename(columns=dict_rrid2gpid, inplace=True)
print('# - Final shape:', df_tmm_cleaned.shape)

# Save to output, bgzip and tabix
# output_path = '/belowshare/vumcshare/data100t1/home/wanying/CCHC/eQTL_gtex_pipeline/input'
output_path = '/vgipiper04/CCHC/RNAseq/batch1_2_3_4/GTEx/expression/expression_for_eQTL_id_mapped/'
output_fn = 'CCHC_batch1_2_3_4_TMM.genotype_id.low_quality_dropped.expression.no_dup.filtered.bed'
if not os.path.isfile(os.path.join(output_path, output_fn)) and not os.path.isfile(os.path.join(output_path, output_fn+'.gz')):
    df_tmm_cleaned.to_csv(os.path.join(output_path, output_fn), sep='\t', index=False)
else: print('\n# Output file exists. Skip saving')
```

# 2. Create ID mapped separate covariate files (pheno, PC, PEER)
Format needed:

```python
# Save list of samples into eQTL mapping
path = '/belowshare/vumcshare/data100t1/home/wanying/CCHC/eQTL_gtex_pipeline/supporting_files'
df_pheno_mapped_cleaned_id_mapped.to_csv(f'{path}/final_list_of_smaples_into_eqtl.txt',
                                         sep='\t', index=False)
```

```python
# Load PCA
fn_pc = '/data100t1/home/wanying/CCHC/doc/genotype_and_pc/202210_CCHC_PCs_id_fixed_1KG_controls_removed.txt'
df_pc = pd.read_csv(fn_pc, sep='\t').T
df_pc.columns = df_pc.iloc[2, :]
df_pc.drop(index=['FID', 'IID', 'RRID', 'LABID'], inplace=True)
print('# PCs:', df_pc.shape)
display(df_pc.head(2))

# Load PEER factors
fn_peer = '/vgipiper04/CCHC/RNAseq/batch1_2_3_4/GTEx/peer_factors/CCHC_batch1_2_3_4.low_quality_dropped.LABID.PEER_covariates.txt'
df_peer = pd.read_csv(fn_peer, sep='\t')
print('# PEER factors:', df_peer.shape)
display(df_peer.head(2))
```

```python
# Conver IDs in PCA and PEER factor to genotype IDs
# df_genotype_id
df_pc.rename(columns=dict_rrid2gpid, inplace=True)
df_peer.rename(columns=dict_labid2rrid, inplace=True)
df_peer.rename(columns=dict_rrid2gpid, inplace=True)

display(df_pc.head(2))
display(df_peer.head(2))

# Filter out samples in cleaned RNAseq data
col_sample_ids = list(df_tmm_cleaned.columns[4:])
df_pc_cleaned = df_pc[col_sample_ids].copy().reset_index().rename(columns={'index':'ID'})
df_peer_cleaned = df_peer[['ID'] + col_sample_ids].copy()
print(df_pc_cleaned.shape, df_peer_cleaned.shape)

# Save cleaned and ID order matched PCA and PEER factor files to output
output_path = '/data100t1/home/wanying/CCHC/eQTL_gtex_pipeline/supporting_files'
out_peer = 'CCHC_batch1_2_3_4.low_quality_dropped.genotype_id.no_dup.cleaned.PEER_covariates.txt'
out_pc = 'CCHC_batch1_2_3_4.low_quality_dropped.genotype_id.no_dup.cleaned.PCA.txt'
if not os.path.isfile(os.path.join(output_path, out_pc)):
    df_pc_cleaned.to_csv(os.path.join(output_path, out_pc), sep='\t', index=False)
else: print('# Cleaned PCA file exists, skip saving')

if not os.path.isfile(os.path.join(output_path, out_peer)):
    df_peer_cleaned.to_csv(os.path.join(output_path, out_peer), sep='\t', index=False)
else: print('# Cleaned PEER factor file exists, skip saving')
```

```python
# Reformat phenotype covariates
df_pheno_reformat = df_pheno_mapped_cleaned_id_mapped.T
df_pheno_reformat.columns = df_pheno_reformat.iloc[-1, :]
df_pheno_reformat.drop(index=['RRID', 'LABID', 'VISIT', 'genotype_ID'], inplace=True)
df_pheno_reformat.reset_index(inplace=True)
df_pheno_reformat.rename(columns={'index':'ID'}, inplace=True)
display(df_pheno_reformat)

output_path = '/data100t1/home/wanying/CCHC/eQTL_gtex_pipeline/supporting_files'
out_pheno = 'CCHC_batch1_2_3_4.low_quality_dropped.genotype_id.no_dup.cleaned.pheno_covariates.txt'
df_pheno_reformat.to_csv(os.path.join(output_path, out_pheno), sep='\t', index=False)
```
