# 1. Nominal run

```python
%%bash
# Example run on chr1
chr=1
n_peers=10
path_script=/data100t1/home/wanying/CCHC/eQTL_gtex_pipeline/code/fastqtl/python
expression=/data100t1/home/wanying/CCHC/eQTL_gtex_pipeline/input/CCHC_batch1_2_3_4_TMM.chr${chr}.bed.gz
vcf=/vgipiper04/CCHC/TOPMed_postimpute_042022/chr${chr}.dose.vcf.gz
prefix=chr${chr}.${n_peers}_peers.nominal
covariates=/data100t1/home/wanying/CCHC/eQTL_gtex_pipeline/supporting_files/combined_covar_sex_age_pc1-5.${n_peers}_peers.combined_covariates.txt

python ${path_script}/run_FastQTL_threaded.py ${vcf} ${expression} ${prefix} \
--covariates ${covariates} \
--window 1e6 \
--ma_sample_threshold 10 \
--maf_threshold 0.01 \
--chunks 100 \
--threads 10
```

# 1. Permutation run
Use 60 and 80 peer factors (the files using eQTL samples only)

```python
# For this run --chunks must be <100 for chr22 (will get error due to no gene found in the chunk)
chr=1
n_peers=60
path_script=/data100t1/home/wanying/CCHC/eQTL_gtex_pipeline/code/fastqtl/python
expression=/data100t1/home/wanying/CCHC/eQTL_gtex_pipeline/input/CCHC_batch1_2_3_4_TMM.chr${chr}.bed.gz
vcf=/vgipiper04/CCHC/TOPMed_postimpute_042022/chr${chr}.dose.vcf.gz
prefix=chr${chr}.${n_peers}_peers.permute
covariates=/data100t1/home/wanying/CCHC/eQTL_gtex_pipeline/supporting_files/combined_covar_sex_age_pc1-5.${n_peers}_peers.combined_covariates.txt

python ${path_script}/run_FastQTL_threaded.py ${vcf} ${expression} ${prefix} \
--covariates ${covariates} \
--permute 1000 10000 --window 1e6 \
--ma_sample_threshold 10 \
--maf_threshold 0.01 \
--chunks 100 \
--threads 10
```

```python
# CMDs for slurm batch run
for n_peers in [60, 80]:
    for chr_num in range(1,23):
        cmd = f'n_peers=60; bash run_fastQTL_permute.sh {chr_num} {n_peers}'
        print(cmd)
```
